```python

def fixed_point_binary(value, integer_bits, fractional_bits):
    # Multiply the value by 2^fractional_bits to shift the decimal point
    int_value = int(round(value * (2**fractional_bits)))
    
    # Convert the integer value to binary
    binary_string = bin(int_value)[2:]
    
    # Pad the binary string with zeros to the left
    return "0" * (integer_bits + fractional_bits - len(binary_string)) + binary_string

# Test the function
binary_string = fixed_point_binary(2.304, 6, 3)
print(binary_string)  # Output: '0100110'
```
